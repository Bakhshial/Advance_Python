# Array Slicing: Accessing Subarrays

```python
#x[start:stop:step]
x=np.arange(10) #0 to 9
#x=np.arange(2,10) # 2 to 9
x
```

```python
print("print 0 to 5:",x[:6])
print("print 2 to 5:",x[2:6])
print("alternative:",x[::2])
print("alternative:",x[1::2])
print("5 after element:",x[5:])
print("reverse element:",x[::-1])
x[5::-2]  # reversed every other from index 5
```

```python
print("two rows,three columns:\n ",x2[:2,:3])
print("\n",x2[:2,2:])

print("\n",x2[:2,2::-1])

print("\n",x2[1:,2:])
```

# Concatenation of arrays

```python
arr1=np.array([1,2,3,4])
arr2=np.array([3,5,7,8,9])
print(np.concatenate([arr1,arr2]))
arr3=np.array([99,98,97])
np.concatenate([arr1,arr2,arr3])
```

```python
grid = np.array([[1, 2, 3],
                 [4, 5, 6]])
np.concatenate([grid,grid])
```

```python
grid = np.array([[1, 2, 3],
                 [4, 5, 6]])
# concatenate along the second axis (zero-indexed)
np.concatenate([grid,grid],axis=1)
```

```python
x = np.array([1, 2, 3])
grid = np.array([[9, 8, 7],
                 [6, 5, 4]])

# vertically stack the arrays
np.vstack([x, grid])
```

```python
y=np.array([[34],[23]])
np.hstack([grid,y])
```

```python
x = [1, 2, 3, 99, 99, 3, 2, 1]
x1, x2, x3 = np.split(x, [3, 5])
print(x1, x2, x3)
```

```python
grid = np.arange(16).reshape((4, 4))
grid
```

```python
upper,lower=np.vsplit(grid,[2])
print("\n",upper)
print("\n",lower)
```

```python
left,right=np.hsplit(grid,[2])
print("\n",left)
print("\n",right)
```
